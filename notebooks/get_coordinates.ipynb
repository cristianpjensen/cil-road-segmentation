```python
from itertools import product
import numpy as np
```

```python
topleft = (26.755085313701056, -80.3270874180188)
bottomright = (25.752544421084227, -80.16354267839812)
```

```python
lats = np.arange(topleft[0], bottomright[0], step=-0.02)
lons = np.arange(topleft[1], bottomright[1], step=0.02)
```

```python
for lat in lats:
    for lon in lons:
        print(f"{lat}, {lon}")
```
